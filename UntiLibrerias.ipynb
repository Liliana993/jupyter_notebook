Funciones con parametros

---



```python
def media(number_1, number_2, number_3):
  valor = (number_1 + number_2 + number_3) /3
  print(valor)
```

```python
media(10,9,7)
```

```python
def media(lista):
  valor = sum(lista)/len(lista)
  print(valor)
```

```python
media([9,8,7,10,6,9,8,6,10])
```

Funciones que deveulven valores

---



```python
def media(lista):
  valor = sum(lista)/len(lista)
  return(valor)
```

```python
media([9,8,7,6,10,9,8,7])
```

```python
resultado = media([9,8,7,6,10,9,8,7])
resultado
```

```python
def media(lista):
  valor = sum(lista)/len(lista)
  return(valor, len(lista))
```

```python
media([9,8,7,6,10,9,8,7])
```

```python
#desempacamos los datos en 2 variables
resultado, n = media([9,8,7,6,10,9,8,7])
```
